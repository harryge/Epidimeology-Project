```python
def county_name(county):
    county = county.strip()
    if (county == 'Tharaka-Nithi'):
        return 'Tharaka Nithi County'
    
    elif (county == 'Elgeyo-Marakwet'):
        return 'Elgeyo Marakwet County'
    
    else:
        return county + ' County'

county_name('Tharaka-Nithi County')

county_name('Elgeyo-Marakwet County')
        
```

```python
df_pop['Orgunit name'] = df_pop['County'].apply(county_name)
df_pop = df_pop.drop('County', axis=1)
df_pop
```

```python
df = pd.read_csv('KHIS Contraceptives 2021 (2).csv')
df
```

```python
col = df.columns[4:16]
col
```

```python
# df1 = df.pivot(index = 'Orgunit name',
#          columns = 'Data',
#          values = '21-Jan')
# df1.reset_index()

# df1['Month'] = 1
# df1

df_list = [] #we'll fill it with dfs and concat all of them later
m = 1

for el in col:
    df_piv = df.pivot(index = 'Orgunit name',
                      columns = 'Data',
                      values = str(el))
    
    df_piv.reset_index()
    
    df_piv['Month'] = m
    df_list.append(df_piv)
    m += 1
```

```python
df_final = pd.concat(df_list)
df_final
```

```python
df_final.columns
```

```python
df_final = df_final.fillna(0)
df_final.head()
```

```python
df_hiv_year = df_hiv.groupby('Orgunit name').sum()
df_hiv_year = df_hiv_year.drop(['Month', 'Infection_rate'], axis=1)
df_hiv_year
```

```python
def infection_rate(pos,total):
    if total == 0:
        return 0
    else:
        ir = pos/total
        return ir
```

```python
df_hiv_year['Infection_rate'] = np.vectorize(infection_rate)(df_hiv_year['MOH 731   Positive_Total   (Sum  HV01-18 to  HV01-25) HV01-26'], df_hiv_year['Total_tested'])
df_hiv_year
```

```python
# df_hiv_year.columns
df_hiv_year['Infection_rate_perc'] = df_hiv_year['Infection_rate'] * 100
df_hiv_year
```

```python
df_year = pd.merge(df_hiv_year,df_year,how='inner',on='Orgunit name')
df_year
```
